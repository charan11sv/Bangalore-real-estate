```python
#imp
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import Firefox
import time

urls = [
    ['Electronic city', 'https://www.nobroker.in/property/rent/bangalore/Electronic%20City?searchParam=W3sibGF0IjoxMi44NDUyMTQ1LCJsb24iOjc3LjY2MDE2OTUsInBsYWNlSWQiOiJDaElKdy1GUWQ0cHNyanNSSGZkYXpnXzhYRW8iLCJwbGFjZU5hbWUiOiJFbGVjdHJvbmljIENpdHkiLCJzaG93TWFwIjpmYWxzZX1d&radius=2.0&sharedAccomodation=0&city=bangalore&locality=Bangalore%20South&isMetro=false'],
    ['Whitefield', 'https://www.nobroker.in/property/rent/bangalore/Whitefield?searchParam=W3sibGF0IjoxMi45Njk4MTk2LCJsb24iOjc3Ljc0OTk3MjEsInBsYWNlSWQiOiJDaElKZ193TlhmTVJyanNSLVJVQjJCS2x6ekEiLCJwbGFjZU5hbWUiOiJXaGl0ZWZpZWxkIiwic2hvd01hcCI6ZmFsc2V9XQ==&radius=2.0&sharedAccomodation=0&city=bangalore&locality=Bangalore%20South&isMetro=false'],
    ['Marathahalli', 'https://www.nobroker.in/property/rent/bangalore/Marathahalli?searchParam=W3sibGF0IjoxMi45NTY5MjQsImxvbiI6NzcuNzAxMTI3LCJwbGFjZUlkIjoiQ2hJSlZ3a2RWYlFUcmpzUkdVa2VmdGVVZUZrIiwicGxhY2VOYW1lIjoiTWFyYXRoYWhhbGxpIiwic2hvd01hcCI6ZmFsc2V9XQ==&radius=2.0&sharedAccomodation=0&city=bangalore&locality=Bangalore%20South&isMetro=false'],
    ['Malleshwaram', 'https://www.nobroker.in/property/rent/bangalore/Malleswaram?searchParam=W3sibGF0IjoxMy4wMDU1MTEzLCJsb24iOjc3LjU2OTIzNTgsInBsYWNlSWQiOiJDaElKdzg5NV8zODlyanNSdnFXa1JucXN1dDgiLCJwbGFjZU5hbWUiOiJNYWxsZXN3YXJhbSIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&sharedAccomodation=0&city=bangalore&locality=Bangalore%20South&isMetro=false'],
    ['Padmanabhanagara', 'https://www.nobroker.in/property/rent/bangalore/Padmanabhanagar?searchParam=W3sibGF0IjoxMi45MTU1NjkxLCJsb24iOjc3LjU1Njc3MzMsInBsYWNlSWQiOiJDaElKUVhzcWNkTV9yanNSLTZabGlqRkN3d2MiLCJwbGFjZU5hbWUiOiJQYWRtYW5hYmhhbmFnYXIiLCJzaG93TWFwIjpmYWxzZX1d&radius=2.0&sharedAccomodation=0&city=bangalore&locality=Bangalore%20South&isMetro=false'],
    ['Gandhinagar', 'https://www.nobroker.in/property/rent/bangalore/Gandhinagar%20Bus%20Stop?searchParam=W3sibGF0IjoxMi45NzY4MDc2LCJsb24iOjc3LjU3OTg0MzQsInBsYWNlSWQiOiJDaElKODN4Q2g5RVhyanNSekxhM0o5VFB2Wk0iLCJwbGFjZU5hbWUiOiJHYW5kaGluYWdhciBCdXMgU3RvcCIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&sharedAccomodation=0&city=bangalore&locality=Bangalore%20South&isMetro=false']
]

def get_property_count(driver):
    count_element = driver.find_element(By.ID, "propertyCount")
    return int(count_element.text.replace(',', ''))

import time
from selenium.webdriver.common.by import By

def scroll_until_property_count(driver, property_count, max_attempts=10):
    SCROLL_PAUSE_TIME = 2.0
    articles = driver.find_elements(By.XPATH, '//article')
    attempts = 0
    
    while len(articles) < property_count:
        print('len of articles', len(articles))
        previous_length = len(articles)
        
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(SCROLL_PAUSE_TIME)
        articles = driver.find_elements(By.XPATH, '//article')
        
        if len(articles) == previous_length:
            attempts += 1
        else:
            attempts = 0
        
        if attempts >= max_attempts:
            print(f"Stopping scroll after {attempts} attempts with no new articles.")
            break
    
    return articles

def get_property_urls(articles, count):
    urls = []
    for article in articles:
        link = article.find_element(By.XPATH, './/a[@href]')
        urls.append(link.get_attribute('href'))
    return urls

def scrape_urls(locations_and_urls, rate_limit=10):
    wait = WebDriverWait(driver, 10)
    
    for entry in locations_and_urls:
        location = entry[0]
        url = entry[1]
        driver.get(url)
        wait.until(EC.presence_of_element_located((By.ID, "propertyCount")))
        
        property_count = get_property_count(driver)
        print(f"Property count found for {location}: {property_count}")
        articles = scroll_until_property_count(driver, property_count)
        print("scroll done")
        urls = get_property_urls(articles, property_count)
        print('len of urls',len(urls))
        for property_url in urls:
            all_property_data.append((location, property_url))
        print('len of cumulative property data',len(all_property_data))
        if rate_limit:
            time.sleep(60 / rate_limit)
    
    driver.quit()
    return all_property_data

# Initialize the webdriver (you may need to modify the path to the driver executable)
# driver = Firefox()

scraped_data = scrape_urls(urls)
for data in scraped_data:
    print(f"Location: {data[0]}, URL: {data[1]}")
```
